### **데이터 EDA**

#범주형 변수 10개
1. Gender : 성별
2. Dependent_count : 부양가족 수 
3. Education_Level : 교육 수준 (고등학교, 대졸 등)
4. Marital_Status : 결혼 상태(기혼,미혼,이혼,불명)
5. Income_Category : 소득 범주 (범위) 
6. Card_Category : 카드 유형(블루, 실버, 골드, 플래티넘) 
7. Months_on_book : 은행과의 관계기간(월기준) 
8. Total_Relationship_Count : 고객이 은행이 보유한 제품를 가진 총수
9. Months_Inactive_12_mon : 지난 12개월 동안 비활성 개월 수 
10. Contacts_Count_12_mon : 지난 12개월 동안의 연락 횟수 


#연속형 변수 9개
1. Credit_Limit : 신용카드 한도 - 연속
2. Total_Revolving_Bal : 신용카드의 총 회전잔액  -  연속 
3. Avg_Open_To_Buy : 신용 한도 구매 가능(최근 12개월 평균) - 연속
4. Total_Amt_Chng_Q4_Q1 : 거래 금액 변경(Q1 대비 Q4) - 연속 
5. Total_Trans_Amt : 총 거래 금액(지난 12개월) - 연속
6. Total_Trans_Ct : 총 거래 건수(지난 12개월) - 연속 
7. Total_Ct_Chng_Q4_Q1: 거래 수의 변화(Q1 대비 Q4) - 연속
8. Avg_Utilization_Ratio : 평균 카드 이용률 - 연속
9. Customer_Age : 고객 나이- 연속

```python
plt.figure(figsize=(10,10))
import matplotlib.colors as mcl
from matplotlib.colors import LinearSegmentedColormap

########################################1#########################################
fig = plt.figure(figsize=(8,8)) ## 캔버스 생성
plt.subplot(3,1,1)
fig.set_facecolor('white') ## 배경색 설정
 
_, bins, patches = plt.hist(card["Credit_Limit"], 20, edgecolor = 'black') ## 히스토그램
  
bin_centers = 0.5*(bins[:-1]+bins[1:]) ## 막대의 위치에 따라서 색 그라디언트를 입힌다.
 
## 막대의 위치를 0~1로 만들어준다.
col = bin_centers - min(bin_centers) 
col /= max(col) 
 
## hsv 색상을 선택한다.
h = 317
v = 100
## hsv 색상을 rgb 색상으로 바꿔준다.
cv = [mcl.hsv_to_rgb((h*1/(2*180),0.2,v/100)),
      mcl.hsv_to_rgb((h*1/(2*180),1,v/100)),
      mcl.hsv_to_rgb((h*1/(2*180),0.2,v/100))]
 
cmap=LinearSegmentedColormap.from_list('field_cmap', cv, N=256,gamma=1)
 
for c, p in zip(col, patches):
    plt.setp(p, "facecolor", cmap(c))
 
plt.title(f'신용카드 한도',fontsize=20)
plt.show()

########################################2#########################################
fig = plt.figure(figsize=(8,8)) ## 캔버스 생성
plt.subplot(3,1,2)
fig.set_facecolor('white') ## 배경색 설정
 
_, bins, patches = plt.hist(card["Total_Revolving_Bal"], 20, edgecolor = 'black') ## 히스토그램
  
bin_centers = 0.5*(bins[:-1]+bins[1:]) ## 막대의 위치에 따라서 색 그라디언트를 입힌다.
 
## 막대의 위치를 0~1로 만들어준다.
col = bin_centers - min(bin_centers) 
col /= max(col) 
 
## hsv 색상을 선택한다.
h = 317
v = 100
## hsv 색상을 rgb 색상으로 바꿔준다.
cv = [mcl.hsv_to_rgb((h*1/(2*180),0.2,v/100)),
      mcl.hsv_to_rgb((h*1/(2*180),1,v/100)),
      mcl.hsv_to_rgb((h*1/(2*180),0.2,v/100))]
 
cmap=LinearSegmentedColormap.from_list('field_cmap', cv, N=256,gamma=1)
 
for c, p in zip(col, patches):
    plt.setp(p, "facecolor", cmap(c))
 
plt.title(f'신용카드의 총 회전잔액',fontsize=20)
plt.show()

#######################################3#########################################
fig = plt.figure(figsize=(8,8)) ## 캔버스 생성
plt.subplot(3,1,3)
fig.set_facecolor('white') ## 배경색 설정
 
_, bins, patches = plt.hist(card["Avg_Open_To_Buy"], 20, edgecolor = 'black') ## 히스토그램
  
bin_centers = 0.5*(bins[:-1]+bins[1:]) ## 막대의 위치에 따라서 색 그라디언트를 입힌다.
 
## 막대의 위치를 0~1로 만들어준다.
col = bin_centers - min(bin_centers) 
col /= max(col) 
 
## hsv 색상을 선택한다.
h = 317
v = 100
## hsv 색상을 rgb 색상으로 바꿔준다.
cv = [mcl.hsv_to_rgb((h*1/(2*180),0.2,v/100)),
      mcl.hsv_to_rgb((h*1/(2*180),1,v/100)),
      mcl.hsv_to_rgb((h*1/(2*180),0.2,v/100))]
 
cmap=LinearSegmentedColormap.from_list('field_cmap', cv, N=256,gamma=1)
 
for c, p in zip(col, patches):
    plt.setp(p, "facecolor", cmap(c))
 
plt.title(f'신용 한도 구매 가능',fontsize=20)
plt.show()
```
